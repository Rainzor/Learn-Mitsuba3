```python
# import mitsuba 
import mitsuba as mi
import drjit as dr
```

```python
mi.variants()
```

```python
# set variant
mi.set_variant("cuda_ad_rgb")
```

```python
# load scene
scene = mi.load_file("../scenes/lego/lego.xml")
```

```python
# render scene
image = mi.render(scene, spp=64)
```

```python
import matplotlib.pyplot as plt

plt.axis("off")
plt.imshow(image ** (1.0 / 2.2)); # approximate sRGB tonemapping
```

```python
mi.util.write_bitmap("lego.png", image)
mi.util.write_bitmap("lego.exr", image)
```
